**Q6** Sum all positive elements

<font size = "3">

Below, we create a random `numpy` array with positive and negative numbers. Using a `for` loop with a `continue` statement, compute the sum of all the **positive** elements of the array

````python
N = 25
random_array = np.random.uniform(low = -5, high = 10, size = N)

# your code here
my_sum = 0 
for idx in range(N):
    next_number = random_array[idx]
    if next_number <= 0:
        continue 
    my_sum = my_sum + next_number 

print("my_sum:", my_sum)
print("expected solution:", np.sum(random_array[random_array > 0]))

# use of continue keyword 
# continue = skip button 
# 
````
